# Filter Data Feb 2024

```python
import pandas as pd
data = pd.read_csv("INPUT_SQL_Text_Data_Astronomy_and_Astrophysics.csv")
```

```python
data
```

```python
# Identify duplicate rows based on 'year', 'title', 'abstract', and 'country' columns
duplicate_rows = data[data.duplicated(subset=['year', 'title', 'abstract', 'country'], keep=False)]
duplicate_rows
```

```python
data_cleaned = data.drop_duplicates(subset=['year', 'title', 'abstract', 'country'], keep='first')
```

```python
data_cleaned
```

```python
# Count the number of rows with at least one null value
num_rows_with_null = data_cleaned.isnull().any(axis=1).sum()
num_rows_with_null
```

```python
# Filter and display rows with at least one null value
rows_with_null = data_cleaned[data_cleaned.isnull().any(axis=1)]
rows_with_null
```

```python
# Remove rows with null values
data_cleaned = data_cleaned.dropna()
```

```python
data_cleaned
```

```python
# Identify rows with empty strings in any column
rows_with_empty_strings = data_cleaned.applymap(lambda x: x == '').any(axis=1)

# Filter and display rows with empty strings
rows_with_empty = data_cleaned[rows_with_empty_strings]
rows_with_empty
```

```python
data_cleaned.to_csv("filtered_data_feb_2024.csv", index=False)
```
