# Abstract Data Citation Count

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('preprocessed_data_Feb_2024.csv')
df
```

```python
incitations = pd.read_csv('INPUT_SQL_Incitations_Astronomy_and_Astrophysics.csv')
incitations
```

```python
# Replace invalid 'year' values with a placeholder and convert to int
incitations['year'] = incitations['year'].replace('__HIVE_DEFAULT_PARTITION__', '-1').astype(int)
```

```python
# Function to parse citations and filter within a five-year timeframe
def filter_citations_within_five_years(row):
    # Check if 'incitations' is a string (which means it's not NaN)
    if isinstance(row['incitations'], str):
        # Split the incitations string into a list
        citations_list = row['incitations'].split(',')
        # Filter and return citations within a five-year timeframe
        filtered_citations = []
        for citation in citations_list:
            # Split each citation into year and citing work ID
            year, citing_work_id = citation.split('_')
            # Ensure year is treated as integer
            year = int(year)
            # Check if the citation is within five years of the publication year
            if 0 <= (year - row['year']) <= 5:
                filtered_citations.append(citation)
        return filtered_citations
    else:
        # Return an empty list if 'incitations' is NaN
        return []
```

```python
# Apply the updated function to each row in the dataframe
incitations['filtered_citations'] = incitations.apply(filter_citations_within_five_years, axis=1)
```

```python
# Display the updated dataframe
print(incitations[['work_id', 'year', 'filtered_citations']])
```

```python
# Calculate the total number of filtered citations for each work
incitations['total_filtered_citations'] = incitations['filtered_citations'].apply(len)
```

```python
incitations
```

```python
# Create a new dataframe with work_id and the total number of filtered citations
citation_counts = incitations[['work_id', 'total_filtered_citations']]
```

```python
# Merge this citation count data back into the original dataframe df
df = pd.merge(df, citation_counts, on='work_id', how='left')
df
```

```python
# Fill NaN values in 'total_filtered_citations' with 0, assuming no citations means zero citations
df['total_filtered_citations'] = df['total_filtered_citations'].fillna(0)
```

```python
# Convert the 'total_filtered_citations' column to integer, as it represents counts
df['total_filtered_citations'] = df['total_filtered_citations'].astype(int)
```

```python
df
```

```python
# Count the number of rows with 0 value in 'total_filtered_citations'
count_zero_citations = (df['total_filtered_citations'] == 0).sum()
count_zero_citations
```

```python
# number of work IDs in df not present in incitations
work_ids_not_in_incitations = len(set(df['work_id']) - set(incitations['work_id']))
work_ids_not_in_incitations
```

# Deflated Citation Count

```python
# Calculate the earliest and latest years in your dataset
start_year = df['year'].min()
end_year = df['year'].max()
```

```python
# Calculate total filtered citations for the start and end years
start_citations = df[df['year'] == start_year]['total_filtered_citations'].sum()
end_citations = df[df['year'] == end_year]['total_filtered_citations'].sum()

# Calculate the number of periods (years) between the start and end years
periods = end_year - start_year

# Define the function to calculate CAGR
def calculate_cagr(end_value, start_value, periods):
    """Calculate the Compound Annual Growth Rate (CAGR)."""
    return (end_value / start_value) ** (1 / periods) - 1

# Calculate CAGR
cagr = calculate_cagr(end_citations, start_citations, periods)

print(f"The Compound Annual Growth Rate (CAGR) for citations from {start_year} to {end_year} is {cagr*100:.2f}%")
```

```python
def calculate_deflated_citations(row, growth_rate, baseline_year=2020):
    years_diff = baseline_year - row['year']
    deflation_index = (1 + growth_rate) ** years_diff
    return row['total_filtered_citations'] / deflation_index
```

```python
growth_rate = 0.1079

# Apply the function to each row
df['Deflated_Citations'] = df.apply(calculate_deflated_citations, args=(growth_rate,), axis=1)
df
```

```python
# add percentage columns
df['citations_percentage'] = (df['total_filtered_citations'] / df['total_filtered_citations'].sum()) * 100
df['deflated_citations_percentage'] = (df['Deflated_Citations'] / df['Deflated_Citations'].sum()) * 100
df
```

```python
df_filtered = df[df['total_filtered_citations'] > 0]
df_filtered
```

```python
print(df_filtered[['work_id', 'year', 'total_filtered_citations', 'Deflated_Citations', 'citations_percentage', 'deflated_citations_percentage']])
```
