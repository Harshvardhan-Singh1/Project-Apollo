```python
t1 = pd.read_csv('Astronomical Observatories.csv')
t2 = pd.read_csv('List of radio telescopes.csv')
t3 = pd.read_csv('List of Solar Telescope Optical.csv')
t4 = pd.read_csv('List of Solar Telescope Radio.csv')
t5 = pd.read_csv('List of space telescopes.csv')
```

```python
# Extract the 'name' column from each dataframe
names_t1 = t1['Name']
names_t2 = t2['Name']
names_t3 = t3['Name/Observatory']
names_t4 = t4['Name/Observatory']
names_t5 = t5['Name']
```

```python
# Concatenate these columns
labels = pd.concat([names_t1, names_t2, names_t3, names_t4, names_t5], ignore_index=True)
labels
```

```python
# Find all duplicate names
duplicate_labels = labels[labels.duplicated(keep=False)]
duplicate_labels
```

```python
# Remove duplicates
labels = labels.drop_duplicates()
labels
```

```python
# Check if all names from each original dataframe are present in the labels
all_t1_present = names_t1.isin(labels).all()
all_t2_present = names_t2.isin(labels).all()
all_t3_present = names_t3.isin(labels).all()
all_t4_present = names_t4.isin(labels).all()
all_t5_present = names_t5.isin(labels).all()

# Print results
print("All names from t1 present in labels:", all_t1_present)
print("All names from t2 present in labels:", all_t2_present)
print("All names from t3 present in labels:", all_t3_present)
print("All names from t4 present in labels:", all_t4_present)
print("All names from t5 present in labels:", all_t5_present)
```

```python
labels.to_csv('Labels.csv', index=False)
```
