# Citation Disruption Index

```python
import pandas as pd

df_incitations = pd.read_csv('INPUT_SQL_Incitations_Astronomy_and_Astrophysics.csv')
df_outcitations = pd.read_csv('INPUT_SQL_Outcitations_Astronomy_and_Astrophysics.csv')
#df_outcitations_of_incitations = pd.read_csv('INPUT_SQL_Outcitations_of_Incitations_Astronomy_and_Astrophysics.csv')
```

```python
def calculate_DI1(work_id, df_incitations, df_outcitations, df_outcitations_of_incitations):

    references = set(df_outcitations[df_outcitations['work_id'] == work_id]['outcitation_id'])
    
    incitations = df_incitations[df_incitations['work_id'] == work_id]['incitation_id']
    
    N_F = 0
    N_B = 0
    
    for incitation_id in incitations:
        incitation_outcitations = set(df_outcitations_of_incitations[df_outcitations_of_incitations['work_id'] == incitation_id]['outcitation_id'])
        
        if incitation_outcitations.isdisjoint(references):
            N_F += 1
        else:
            N_B += 1
    
    N_R = len(references)
    
    DI1 = (N_F - N_B) / (N_F + N_B + N_R) if (N_F + N_B + N_R) > 0 else None
    
    return DI1


df_incitations['DI1'] = df_incitations['work_id'].apply(lambda x: calculate_DI1(x, df_incitations, df_outcitations, df_outcitations_of_incitations))

print(df_incitations[['work_id', 'DI1']])

```
