```python
import pandas as pd
file_path = "Astronomy_Constellate_FullText_Metadata_JSONL"
```

```python

data = pd.read_json(file_path, lines=True)


```

```python
data
```

```python
for column in data.columns:
        num_nan = data[column].isna().sum()
        print(f"There are {num_nan} NaN values in the '{column}' column.")
```

```python
# Extract year and create a new column 'publication_year'
data['publication_year'] = data['datePublished'].str.split('-').str[0]
data
```

```python
earliest_year = data['publication_year'].min()
latest_year = data['publication_year'].max()

print(f"Earliest year of publication: {earliest_year}")
print(f"Latest year of publication: {latest_year}")
```

```python
# Convert the publication_year column to integer for comparison
data['publication_year'] = data['publication_year'].astype(int)

# Define the intervals
intervals = {
    "pub_interval_2020_2022": (2020, 2022),
    "pub_interval_2015_2019": (2015, 2019),
    "pub_interval_2010_2014": (2010, 2014),
    "pub_interval_2005_2009": (2005, 2009),
    "pub_interval_2000_2004": (2000, 2004),
    "pub_interval_1995_1999": (1995, 1999),
    "pub_interval_1985_1994": (1985, 1994),
    "pub_interval_1975_1984": (1975, 1984),
    "pub_interval_1965_1974": (1965, 1974),
    "pub_interval_1900_1964": (1900, 1964),
    "pub_interval_1824_1899": (1824, 1899)
}

# Populate the interval columns
for column_name, (start_year, end_year) in intervals.items():
    data[column_name] = data['publication_year'].apply(lambda year: 1 if start_year <= year <= end_year else 0)
```

```python
data.to_csv('preprocessed_FullText_data.csv', index=False)
```
